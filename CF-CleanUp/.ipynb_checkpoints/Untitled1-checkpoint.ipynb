```python
import pandas as pd
import numpy as np
from fuzzywuzzy import process

sf = pd.read_excel('B8BB5200.xlsx',sheet_name='5')
sf['Account.Axcess ID'] = sf['Account.Axcess ID'].fillna('')
cch = pd.read_csv('CCH_CF_Client.csv')
# pb = pd.read_csv('data.csv')
# sf['Split_Product']=sf['Split_Product'].str.replace('Acct Service','Accounting Services')
# sf['Opportunity Name'].str.split('-',expand=True)[0]
# sf['Split_EntityName']=sf['Opportunity Name'].str.split('-',expand=True)[0]
# sf['Split_PriceBook']=sf['Opportunity Name'].str.split('-',expand=True)[1].str.replace('Acct Service','Accounting Services').str.lower()
# sf['Split_Product']=sf['Opportunity Name'].str.split('-',expand=True)[2]

cch['ClientNameSplit']=cch['ClientSortName'].str.split('-',expand=True)[0]
# cch['ProductSplit']=cch['ClientSortName'].str.split('-',expand=True)[1].str.replace('AS','Accounting Services').str.lower()
cch['ClientNameFull'] = cch['ClientIdSubId']+' - '+cch['ClientSortName']
cch['ClientNameFull1'] = cch['ClientIdSubId']+' - '+cch['CorrespondenceName']
# test = pb['Family'].unique()
# for row in test:
#     pb.loc[len(pb.index)] = [row, row]
# pb.loc[len(pb.index)] = ['AS', 'Accounting Services']
# pb['Name'] =pb['Name'].str.lower()


D= {}
D2 = {}
D3 = {}
D31 = {}
D4 = {}
i=0
for row in sf.index:
    temp_cch = cch[(cch['ClientId']==sf.at[row,'Account.Axcess ID'])]
    temp_sf = [
        sf.at[row,'Legal Entity Name 1']
        ,sf.at[row,'Legal Entity Name 2']
        ,sf.at[row,'Legal Entity Name 3']
        ,sf.at[row,'Legal Entity Name 4']
        ,sf.at[row,'Legal Entity Name 5']
        ,sf.at[row,'Legal Entity Name 6']
        ,sf.at[row,'Legal Entity Name 7']
        ,sf.at[row,'Legal Entity Name 8']
        ,sf.at[row,'Legal Entity Name 9']
        ,sf.at[row,'Legal Entity Name 10']
    ]
    new_list = [item if item is not np.nan else '' for item in temp_sf]
    D3[sf.at[row,'Opportunity Id']] = [[],[],[]] # legalentity, clientnamefull, % match
    D31[sf.at[row,'Opportunity Id']] = [[],[],[]] # legalentity, clientnamefull, % match
    if str(sf.at[row,'Account.Axcess ID']) == '':
        D[sf.at[row,'Opportunity Id']] = 'No Axcess Code in Account'
    elif len(temp_cch.index)==0:
        D[sf.at[row,'Opportunity Id']] = 'Axcess Code not found in CCH'
    elif len(temp_cch.index)>0:
        for j in temp_cch.index:
            #fuzzy match ClientSortName
            best_match, score = process.extractOne(temp_cch.at[j,'ClientNameSplit'], new_list)
            D3[sf.at[row,'Opportunity Id']][0].append(best_match)
            D3[sf.at[row,'Opportunity Id']][1].append(temp_cch.at[j,'ClientNameFull'])
            D3[sf.at[row,'Opportunity Id']][2].append(score)
            #fuzzy match CorrespondenceName
            best_match1, score1 = process.extractOne(temp_cch.at[j,'CorrespondenceName'], new_list)
            D31[sf.at[row,'Opportunity Id']][0].append(best_match1)
            D31[sf.at[row,'Opportunity Id']][1].append(temp_cch.at[j,'ClientNameFull1'])
            D31[sf.at[row,'Opportunity Id']][2].append(score1)
            
            #match absolute 
            if sf.at[row,'Legal Entity Name 1'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull']
                break
            elif sf.at[row,'Legal Entity Name 2'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull']      
                break
            elif sf.at[row,'Legal Entity Name 3'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull']
                break
            elif sf.at[row,'Legal Entity Name 4'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull']
                break
            elif sf.at[row,'Legal Entity Name 5'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull']
                break
            elif sf.at[row,'Legal Entity Name 6'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull']    
                break
            elif sf.at[row,'Legal Entity Name 7'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull']
                break
            elif sf.at[row,'Legal Entity Name 8'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull']    
                break
            elif sf.at[row,'Legal Entity Name 9'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull']
                break
            elif sf.at[row,'Legal Entity Name 10'] == temp_cch.at[j,'ClientNameSplit']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull'] 
                break
                
            #match absolute     
            elif sf.at[row,'Legal Entity Name 2'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1']      
                break                
            elif sf.at[row,'Legal Entity Name 2'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1']      
                break
            elif sf.at[row,'Legal Entity Name 3'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1']
                break
            elif sf.at[row,'Legal Entity Name 4'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1']
                break
            elif sf.at[row,'Legal Entity Name 5'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1']
                break
            elif sf.at[row,'Legal Entity Name 6'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1']    
                break
            elif sf.at[row,'Legal Entity Name 7'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1']
                break
            elif sf.at[row,'Legal Entity Name 8'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1']    
                break
            elif sf.at[row,'Legal Entity Name 9'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1']
                break
            elif sf.at[row,'Legal Entity Name 10'] == temp_cch.at[j,'CorrespondenceName']:
                D2[sf.at[row,'Opportunity Id']] = temp_cch.at[j,'ClientNameFull1'] 
                break
        numbers_array = D3[sf.at[row,'Opportunity Id']][2]
        max_value = max(numbers_array)
        
        numbers_array1 = D31[sf.at[row,'Opportunity Id']][2]
        max_value1 = max(numbers_array1)
        if max_value >= max_value1:
            max_index = numbers_array.index(max_value)
            LegalEntity = D3[sf.at[row,'Opportunity Id']][0][max_index]
            ClientNameFull = D3[sf.at[row,'Opportunity Id']][1][max_index]
            D4[sf.at[row,'Opportunity Id']] = [LegalEntity, ClientNameFull, max_value,'ClientSortName']
        else:
            max_index1 = numbers_array1.index(max_value1)
            LegalEntity1 = D31[sf.at[row,'Opportunity Id']][0][max_index1]
            ClientNameFull1 = D31[sf.at[row,'Opportunity Id']][1][max_index1]
            D4[sf.at[row,'Opportunity Id']] = [LegalEntity1, ClientNameFull1, max_value1, 'CorrespondenceName']
        i=i+1
#         D[sf.at[row,'Opportunity Id']] = str(list(temp_cch['ClientNameFull'])).replace('[','').replace(']','').replace("''",'')
        D[sf.at[row,'Opportunity Id']] = '|'.join(list(temp_cch['ClientNameFull']))

    
    
Fuzzy = pd.DataFrame.from_dict(D4, orient='index')  
Fuzzy.reset_index(inplace=True)
Fuzzy.rename(columns={'index': 'CCH3.Opportunity ID', 0: 'Fuzzy.Legal Entity', 1: 'Fuzzy.ClientNameFull', 2: 'Fuzzy.% match', 3: 'Fuzzy.NameType'}, inplace=True)
ExMatch = pd.DataFrame.from_dict(D2, orient='index')  
ExMatch.reset_index(inplace=True)
ExMatch.rename(columns={'index': 'CCH2.Opportunity ID', 0: 'Absolute match'}, inplace=True)
FullC = pd.DataFrame.from_dict(D, orient='index')  
FullC.reset_index(inplace=True)
FullC.rename(columns={'index': 'CCH1.Opportunity ID', 0: 'Suggested CCH Client'}, inplace=True)
FullC.head()
merged1 = pd.merge(sf, Fuzzy, how='left', left_on='Opportunity Id', right_on='CCH3.Opportunity ID')
merged2 = pd.merge(merged1, ExMatch, how='left', left_on='Opportunity Id', right_on='CCH2.Opportunity ID')
merged3 = pd.merge(merged2, FullC, how='left', left_on='Opportunity Id', right_on='CCH1.Opportunity ID')
merged3.head()
#choose columns to export
merged3.columns

result_pd = merged3[['Opportunity Id','Account.Axcess ID'
                   ,'Legal Entity Name 1','Legal Entity Name 2','Legal Entity Name 3','Legal Entity Name 4','Legal Entity Name 5'
                   ,'Legal Entity Name 6','Legal Entity Name 7','Legal Entity Name 8','Legal Entity Name 9','Legal Entity Name 10'
                   ,'Suggested CCH Client', 'Absolute match', 'Fuzzy.Legal Entity', 'Fuzzy.ClientNameFull', 'Fuzzy.% match',
       'Fuzzy.NameType' ]
                   ]
result_pd.to_csv('ressult_5.csv')
```
