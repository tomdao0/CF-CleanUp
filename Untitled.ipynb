```python
def is_valid_move(x, y, board, N):
    if 0 <= x < N and 0 <= y < N and board[x][y] == -1:
        return True
    return False

def print_solution(board):
    for row in board:
        print(' '.join(map(str, row)))

def solve_knights_tour(N):
    board = [[-1 for _ in range(N)] for _ in range(N)]

    # Possible moves for a knight
    move_x = [2, 1, -1, -2, -2, -1, 1, 2]
    move_y = [1, 2, 2, 1, -1, -2, -2, -1]

    # Starting position
    board[0][0] = 0

    if not solve_knights_tour_util(0, 0, 1, board, move_x, move_y, N):
        print("No solution exists")
    else:
        print_solution(board)

def solve_knights_tour_util(x, y, move_num, board, move_x, move_y, N):
    if move_num == N ** 2:
        return True

    for k in range(8):
        next_x = x + move_x[k]
        next_y = y + move_y[k]

        if is_valid_move(next_x, next_y, board, N):
            board[next_x][next_y] = move_num
            if solve_knights_tour_util(next_x, next_y, move_num + 1, board, move_x, move_y, N):
                return True

            # Backtrack
            board[next_x][next_y] = -1

    return False

# N is the size of the board
N = 8
solve_knights_tour(N)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu mẫu: hình sin tần số 50 Hz kết hợp với nhiễu
fs = 1000  # Tần số lấy mẫu
t = np.arange(0, 1, 1/fs)  # Thời gian từ 0 đến 1 giây
f = 50  # Tần số của hình sin (Hz)
x = 0.7 * np.sin(2 * np.pi * f * t) + 0.3 * np.random.normal(size=len(t))

# Thực hiện FFT
X = np.fft.fft(x)
freqs = np.fft.fftfreq(len(x), 1/fs)  # Tần số tương ứng

# Vẽ biểu đồ tín hiệu gốc và biểu đồ tần số
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.xlabel('Thời gian (s)')
plt.ylabel('Amplitude')
plt.title('Tín hiệu gốc')

plt.subplot(2, 1, 2)
plt.plot(freqs, np.abs(X))
plt.xlabel('Tần số (Hz)')
plt.ylabel('Amplitude')
plt.title('Biểu đồ tần số (FFT)')
plt.xlim(0, fs/2)  # Hiển thị chỉ tần số dương
plt.tight_layout()

plt.show()
```

```python
def longest_increasing_subsequence(arr):
    n = len(arr)
    lis = [1] * n  # Khởi tạo mảng LIS với giá trị 1

    for i in range(1, n):
        for j in range(0, i):
            if arr[i] > arr[j] and lis[i] < lis[j] + 1:
                lis[i] = lis[j] + 1

    max_length = max(lis)
    max_index = lis.index(max_length)
    longest_subsequence = []

    for i in range(max_index, -1, -1):
        if lis[i] == max_length:
            longest_subsequence.append(arr[i])
            max_length -= 1

    return longest_subsequence[::-1]

# Dãy số mẫu
sequence = [10, 22, 9, 33, 21, 50, 41, 60, 80]

result = longest_increasing_subsequence(sequence)
print("Dãy con tăng dài nhất:", result)
```

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.hash_table = [None] * self.size

    def hash_function(self, key):
        return sum(ord(char) for char in key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        if self.hash_table[index] is None:
            self.hash_table[index] = [(key, value)]
        else:
            for i, (existing_key, existing_value) in enumerate(self.hash_table[index]):
                if existing_key == key:
                    self.hash_table[index][i] = (key, value)
                    break
            else:
                self.hash_table[index].append((key, value))

    def get(self, key):
        index = self.hash_function(key)
        if self.hash_table[index] is not None:
            for existing_key, existing_value in self.hash_table[index]:
                if existing_key == key:
                    return existing_value
        return None

# Tạo bảng băm có kích thước 10
hash_table = HashTable(10)

# Thêm dữ liệu vào bảng băm
hash_table.insert("apple", 5)
hash_table.insert("banana", 7)
hash_table.insert("cherry", 10)

# Truy vấn dữ liệu từ bảng băm
print("Value for 'apple':", hash_table.get("apple"))
print("Value for 'banana':", hash_table.get("banana"))
print("Value for 'cherry':", hash_table.get("cherry"))
print("Value for 'grape':", hash_table.get("grape"))
```
